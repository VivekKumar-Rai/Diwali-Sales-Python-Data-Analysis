## Exploratory Data Analysis

### Gender

```python
df.columns
```

```python
a = sn.countplot(x = 'Gender', data = df)

sn.set(rc={'figure.figsize':(6,5)})

for labels in a.containers:
    a.bar_label(labels)
```

```python
#Gender vs Amount

sales_gender = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sn.barplot(x = 'Gender', y = 'Amount' , data = sales_gender)
```

###### We can clearly see that Females are the major customers in the total sales.

### Age

```python
df.columns
```

```python
b = sn.countplot(data = df, x = 'Age Group', hue = 'Gender')

for labels in b.containers:
    b.bar_label(labels)
```

```python
#Age Group vs Amount

sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sn.barplot(x = 'Age Group', y = 'Amount' , data = sales_age)
```

#### We can conclude that most of the customers are in the age group of 26-35.

### State

```python
df.columns
```

```python
#Orders vs State

orders_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', 
                                                                                   ascending = False).head(10)

sn.set(rc={'figure.figsize':(18,5)})

sn.barplot(x = 'State', y = 'Orders' , data = orders_state)
```

```python
#Amount vs States

sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', 
                                                                                  ascending = False).head(10)

sn.set(rc={'figure.figsize':(18,5)})

sn.barplot(x = 'State', y = 'Amount' , data = sales_state)
```

#### Hence, here we can conclude that most orders/sales are from Uttar Pradesh, Maharastra and Karnataka as the Top 3 States.

### Marital Status

```python
df.columns
```

```python
df['Marital_Status'] = df['Marital_Status'].astype('string')

df['Marital_Status'] = df['Marital_Status'].replace(to_replace = '0', value = 'Married')
df['Marital_Status'] = df['Marital_Status'].replace(to_replace = '1', value = 'Unmarried')
```

```python
c = sn.countplot(data = df, x = 'Marital_Status', hue = 'Gender')

sn.set(rc={'figure.figsize':(7,5)})

for labels in c.containers:
    c.bar_label(labels)
```

```python
sales_status = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', 
                                                                                                    ascending=False)

sn.set(rc={'figure.figsize':(6,5)})

sn.barplot(data = sales_status, x = 'Marital_Status', y= 'Amount', hue='Gender')
```

#### Hence, from here we can clearly see that Married Females provide the highest sales/orders.

### Occupation

```python
#Occupation vs Orders

d = sn.countplot(data = df, x = 'Occupation')

sn.set(rc={'figure.figsize':(20,5)})

for labesl in d.containers:
    d.bar_label(labels)
```

```python
#Occupation vs Sales

sales_occupation = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', 
                                                                                          ascending=False)

sn.set(rc={'figure.figsize':(20,5)})

sn.barplot(data = sales_occupation, x = 'Occupation', y= 'Amount')
```

#### From these graphs we can understand that most of the customers are working in the IT Sector, Healthcare, Aviation and Banking.

### Product Category

```python
d = sn.countplot(data = df, x = 'Product_Category')

sn.set(rc={'figure.figsize':(30,5)})

for labels in d.containers:
    d.bar_label(labels)
```

```python
sales_product = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', 
                                                                                             ascending=False).head(10)

sn.set(rc={'figure.figsize':(25,5)})

sn.barplot(data = sales_product, x = 'Product_Category', y = 'Amount')
```

#### Hence, here we can conclude that the highest sales is of Food followed by Clothing & Apparel, Electronics & Gadgets and Footwear & Shoes. 

```python
sales_product_id = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', 
                                                                                          ascending=False).head(10)

sn.set(rc={'figure.figsize':(20,5)})

sn.barplot(data = sales_product_id, x = 'Product_ID', y= 'Orders')
```

```python
sales = df.groupby(['Product_ID', 'Product_Category'], as_index=False)['Orders'].sum().sort_values(by='Orders', 
                                                                                                   ascending=False).head(10)

sales
```

#### From the above table and graph we can see the exact product and their category with the highest sales.
